```python
income_mapping = {
    'Less than 10,000': 10000,
    '10,000 to 14,999': 14999,
    '15,000 to 24,999': 24999,
    '25,000 to 34,999': 34999,
    '35,000 to 49,999': 49999,
    '50,000 to 74,999': 74999,
    '75,000 to 99,999': 99999,
    '100,000 to 149,999': 149999,
    '150,000 to 199,999': 199999,
    '200,000 or more': 300000
}

pop_columns = ['Less than 10,000', '10,000 to 14,999', '15,000 to 24,999', '25,000 to 34,999',
               '35,000 to 49,999', '50,000 to 74,999', '75,000 to 99,999', '100,000 to 149,999',
               '150,000 to 199,999', '200,000 or more']

income_columns = ['Less than 10,000', '10,000 to 14,999', '15,000 to 24,999', '25,000 to 34,999',
                   '35,000 to 49,999', '50,000 to 74,999', '75,000 to 99,999', '100,000 to 149,999',
                   '150,000 to 199,999', '200,000 or more']

df_income_x_city['average_income'] = df_income_x_city.apply(lambda row: sum(row[pop] * income_mapping[inc] for pop, inc in zip(pop_columns, income_columns)) / row['Total Households'] if row['Total Households'] != 0 else 0, axis=1)

df_income_x_city[['City', 'average_income']]
```

```python
df_pivot = df_income_x_city.pivot(index='City', columns='Year', values='average_income')

df_pivot.columns = [f'Average Income {year}' for year in df_pivot.columns]

df_pivot.reset_index(inplace=True)

df_pivot.head()
```

```python
merged_df = pd.merge(df_pop_price, df_pivot, on='City', how='inner')

income_columns = [f'Average Income {year}' for year in range(2015, 2022)]
price_columns = [f'{year} Price' for year in range(2015, 2022)]

years = []
correlation_coefficients = []

for income_col, price_col in zip(income_columns, price_columns):
    years.append(income_col[-4:])
    correlation_coefficients.append(merged_df[income_col].corr(merged_df[price_col]))

correlation_results = pd.DataFrame({'Year': years, 'Correlation Coefficient': correlation_coefficients})

print(correlation_results)
```

```python
import matplotlib.pyplot as plt

plt.plot(correlation_results['Year'], correlation_results['Correlation Coefficient'], marker='o', linestyle='-')
plt.title('Correlation Coefficients Over the Years')
plt.xlabel('Year')
plt.ylim((0.6,0.7))
plt.ylabel('Correlation Coefficient')
plt.grid(True)
plt.show()
```
