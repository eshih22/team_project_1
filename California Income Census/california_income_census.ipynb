```python
import pandas as pd
```

```python
years_files = [str(i)+'.csv' for i in range(2013,2023)]
years = [str(i) for i in range(2015,2022)]
years, years_files
```

```python
tables = dict()

for i in range(len(years)):
    tables[years[i]] = pd.read_csv(years_files[i])
    tables[years[i]]['Year'] = years[i]
    
for i in range(len(years)):
    df = tables[years[i]]
    df.iloc[0, [1,3,5,7] ]= df.iloc[0, [1,3,5,7]].str.replace(',', '').astype(float)
    df.iloc[1:11, [1, 3, 5, 7]] = df.iloc[1:11, [1, 3, 5, 7]].apply(lambda x: x.str.strip('%').astype(float) / 100)
    df.iloc[1:11, [1, 3, 5, 7]] = df.iloc[1:11, [1, 3, 5, 7]].mul(df.iloc[0, [1, 3, 5, 7]])
    df.iloc[1:11, [1, 3, 5, 7]] = df.iloc[1:11, [1, 3, 5, 7]].round(0).astype(int)
    tables[years[i]] = df.drop(index=[13,14,15,16], columns=df.columns[[2,4,6,8]])
    
    df = tables[years[i]]
    df.iloc[[11,12],1:5] = df.iloc[[11,12],1:5].applymap(lambda x: str(x).replace(',', ''))
    df.iloc[0,0] = "Total"
    df.iloc[1,0] = "Less than 10,000"
    df.iloc[2,0] = "10,000 to 14,999"
    df.iloc[3,0] = "15,000 to 24,999"
    df.iloc[4,0] = "25,000 to 34,999"
    df.iloc[5,0] = "35,000 to 49,999"
    df.iloc[6,0] = "50,000 to 74,999"
    df.iloc[7,0] = "75,000 to 99,999"
    df.iloc[8,0] = "100,000 to 149,999"
    df.iloc[9,0] = "150,000 to 199,999"
    df.iloc[10,0] = "200,000 or more"
    df.iloc[11,0] = "Median income (dollars)"
    df.iloc[12,0] = "Mean income (dollars)"
```

```python
tables['2021']
```

```python
megatable = tables[years[0]]

for i in range(1, len(years)):
    df = tables[years[i]]
    megatable = pd.concat([megatable, df], axis=0, ignore_index=True)
    
megatable
```

```python
megatable.to_csv("california_incomes_2015_2021.csv", index=False, encoding='utf-8')
```
